```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies = pd.read_csv("movie_metadata.csv")
```

```python
movies.head()
```

```python
movies.info()
```

*** Data preprocessing ***



#1 Let's remove some useless features.

```python
sns.lmplot(y='imdb_score',x='movie_facebook_likes',data=movies)
```

```python
sns.lmplot(y='imdb_score',x='director_facebook_likes',data=movies)
```

```python
sns.lmplot(y='imdb_score',x='cast_total_facebook_likes',data=movies)
```

from the above plots, we can say there is no correkation between the movie imdb ratings an facebook likes! So we will create a new dataset dropping the 'facebook' columns 

```python
fbookcols = movies.columns[movies.columns.str.contains('facebook')]
movie_temp = movies.drop(fbookcols,axis=1)
movie_2 = movie_temp.drop('movie_imdb_link',axis=1)

movie_new_dataset = movie_2[movie_2['num_user_for_reviews'] > 100]
```

```python
movie_new_dataset.info()
```

There are some NaNs let's see how much 

```python
plt.figure(figsize=(13, 40))
plt.rcParams['axes.facecolor'] = 'black'
plt.rc('grid', color='#202020')
(movie_new_dataset.isnull().mean(axis=0)*100).plot.barh(color ="#FF6600")
plt.xlim(xmax=100)
plt.title("Missing values rate",fontsize=18)
plt.xlabel("percentage",fontsize=14)
```
